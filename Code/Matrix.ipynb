```python
# Матрица профиля
import numpy as np

def num(char):
    if char == 'A':
        return 0
    if char == 'C':
        return 1
    if char == 'G':
        return 2
    if char == 'T':
        return 3
    else:
        return 
    
lines = open('input.txt', 'r').readlines()
matrix = np.zeros((4, len(lines)))
for line in enumerate(lines):
    for symbol in line[1].strip():
        matrix[num(symbol)][line[0]] += 1
print(matrix)
```

```python
# лист решений

def rev_num(s):
    if s == 0:
        return 'A'
    if s == 1:
        return 'C'
    if s == 2:
        return 'G'
    if s == 3:
        return 'T'
    else:
        return

answer = []
for i in range(len(lines)):
    max = 0
    ans = ''
    for j in range(4):
        if matrix[j][i] > max:
            max = matrix[j][i]
            ans = rev_num(j)
    answer.append(ans)
print(answer)
```
